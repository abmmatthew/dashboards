Note: date-hour level means that you have to find the count for every hour of the day for each day. Hint: you may need to precess data format in two datasets before merge.


```python
import pandas as pd

# Step 1: Load the datasets
TaxiRidership = pd.read_csv("https://raw.githubusercontent.com/CUSP2020PUI/Data/master/LGAVehicleByHour.csv")
FlightArrival = pd.read_csv("https://raw.githubusercontent.com/CUSP2020PUI/Data/master/LGAFlightArrival.csv")

# Step 2: Convert `Date` columns to datetime format for both datasets
TaxiRidership['Date'] = pd.to_datetime(TaxiRidership['Date'])
FlightArrival['Date'] = pd.to_datetime(FlightArrival['Date'], format='%y/%m/%d %H:%M')

# Step 3: Aggregate Taxi Ridership data to date-hour level
# Extract hour and create a new column combining date and hour
TaxiRidership['date_hour'] = TaxiRidership['Date'].dt.floor('H')

# Sum vehicle_count for each date-hour across all drop-off locations
aggregated_taxi = TaxiRidership.groupby('date_hour')['vehicle_count'].sum().reset_index()

# Step 4: Aggregate Flight Arrival data to date-hour level
# Ensure the `Date` column is already at date-hour level
FlightArrival['date_hour'] = FlightArrival['Date']

# Sum arrivals for each date-hour
aggregated_flights = FlightArrival.groupby('date_hour')['arrivals'].sum().reset_index()

# Step 5: Merge the two datasets on date-hour
merged_data = pd.merge(aggregated_taxi, aggregated_flights, on='date_hour', how='inner')

# Step 6: Preview the resulting dataset
print(merged_data.head())

# Step 7: Save the merged dataset for further analysis
merged_data.to_csv("Merged_Taxi_Flight_Data.csv", index=False)
```


Q2. add time-lagged arrivals (1,2,3,4,5,6..12 hours before) - 5pts

Taxi and FHV ridership volume is impacted by flight arrivals. Intuitively, more flights bring more passengers, and more ridership eventually. However, the hourly ridership volume may not only be related to flight arrivals in the same hour, arrivals 1 hour, 2 hours ago could impact ridership as well, as travelers need some time to pass the customs or wait for luggage. Please add lag of arrivals as additional features, and name new features as lag1 for arrivals 1 hour ago, and lag2 for 2 hours before and so on. You may use the shift function in pandas to add arrivals lag values. For example, df['arrivals'].shift(1) means adding 1 hour lag value of arrivals. iloc function works as well.

```python
import pandas as pd

# Step 1: Load the datasets
TaxiRidership = pd.read_csv("https://raw.githubusercontent.com/CUSP2020PUI/Data/master/LGAVehicleByHour.csv")
FlightArrival = pd.read_csv("https://raw.githubusercontent.com/CUSP2020PUI/Data/master/LGAFlightArrival.csv")

# Step 2: Convert `Date` columns to datetime format for both datasets
TaxiRidership['Date'] = pd.to_datetime(TaxiRidership['Date'])
FlightArrival['Date'] = pd.to_datetime(FlightArrival['Date'], format='%y/%m/%d %H:%M')

# Step 3: Aggregate Taxi Ridership data to date-hour level
# Extract hour and create a new column combining date and hour
TaxiRidership['Hour'] = TaxiRidership['Date'].dt.hour
TaxiRidership['DOW'] = TaxiRidership['Date'].dt.weekday
TaxiRidership['day'] = (TaxiRidership['Date'] - pd.Timestamp('2018-01-01')).dt.days
TaxiRidership['date_hour'] = TaxiRidership['Date'].dt.floor('H')

# Sum vehicle_count for each date-hour across all drop-off locations
aggregated_taxi = TaxiRidership.groupby('date_hour')['vehicle_count'].sum().reset_index()

# Step 4: Aggregate Flight Arrival data to date-hour level
FlightArrival['date_hour'] = FlightArrival['Date']
aggregated_flights = FlightArrival.groupby('date_hour')['arrivals'].sum().reset_index()

# Step 5: Merge the two datasets on date-hour
merged_data = pd.merge(aggregated_taxi, aggregated_flights, on='date_hour', how='inner')

# Step 6: Add lagged arrivals (1 to 12 hours)
for lag in range(1, 13):  # Create lagged features from 1 to 12 hours
    merged_data[f'lag{lag}'] = merged_data['arrivals'].shift(lag)

# Add other relevant columns for output structure
merged_data['Date'] = merged_data['date_hour']
merged_data['Hour'] = merged_data['date_hour'].dt.hour
merged_data['DOW'] = merged_data['date_hour'].dt.weekday
merged_data['day'] = (merged_data['date_hour'] - pd.Timestamp('2018-01-01')).dt.days

# Step 7: Handle missing values for lag features
merged_data.fillna(0, inplace=True)

# Step 8: Rearrange columns for final output structure
final_columns = ['Date', 'Hour', 'DOW', 'day', 'vehicle_count', 'arrivals'] + [f'lag{lag}' for lag in range(1, 13)]
final_dataset = merged_data[final_columns]

# Step 9: Display the first 20 rows of the dataset
print(final_dataset.iloc[:20])

# Step 10: Save the updated dataset for further analysis
final_dataset.to_csv("Final_Dataset_With_Lags.csv", index=False)
```

Q2 (10pts). Plot a probability distribution function of hourly ridership volume on weekdays and weekend, and apply t-test to check is there a significant difference of hourly ridership between each day of week and the rest of the days,
Report hypothesis testing result in a
table, index name should be day of week (e.g. Monday), values should be True or False (True for P < 0.05).

Hint: You may use norm.pdf function in scipy.stats to calculate the probability distribution, and standardize at first.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, ttest_ind

# Separate data into weekdays (DOW 0-4) and weekends (DOW 5-6)
weekdays = final_dataset[final_dataset['DOW'] <= 4]['vehicle_count']
weekends = final_dataset[final_dataset['DOW'] >= 5]['vehicle_count']

# Standardize the ridership data
weekday_std = (weekdays - weekdays.mean()) / weekdays.std()
weekend_std = (weekends - weekends.mean()) / weekends.std()

# Plot the Probability Distribution Function (PDF)
x = np.linspace(-3, 3, 100)
plt.figure(figsize=(10, 6))
plt.plot(x, norm.pdf(x), label='Normal Distribution')
plt.hist(weekday_std, bins=30, density=True, alpha=0.6, label='Weekdays')
plt.hist(weekend_std, bins=30, density=True, alpha=0.6, label='Weekends')
plt.title("Probability Distribution of Hourly Ridership (Weekdays vs Weekends)")
plt.xlabel("Standardized Hourly Ridership")
plt.ylabel("Probability")
plt.legend()
plt.grid()
plt.show()

# Perform t-test for each day of the week
t_test_results = {}
for day in range(7):
    current_day = final_dataset[final_dataset['DOW'] == day]['vehicle_count']
    other_days = final_dataset[final_dataset['DOW'] != day]['vehicle_count']
    t_stat, p_value = ttest_ind(current_day, other_days, equal_var=False)
    t_test_results[day] = p_value < 0.05  # True if p < 0.05

# Report results in a table
t_test_df = pd.DataFrame.from_dict(t_test_results, orient='index', columns=['Significant Difference'])
t_test_df.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
print(t_test_df)

# Save the table to a CSV file
t_test_df.to_csv("T_Test_Results.csv", index=True)
```


Q2. Build an OLS model

Build an OLS model with intercept (you may want to use smf.ols) over the training dataset using arrivals as a sole predictor for vehicle_count. Check p-value for arrivals. What does it indicate? Report the 99% confidence interval for arrivals's coefficient. Scatter plot for ridership and flight arrival, include the predictive line.


```python
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import numpy as np

# Build the OLS model
ols_model = smf.ols(formula='vehicle_count ~ arrivals', data=training_data).fit()

# Print the model summary
print(ols_model.summary())

# Extract p-value for arrivals
p_value = ols_model.pvalues['arrivals']
print(f"P-value for arrivals: {p_value}")

# Report 99% confidence interval for arrivals' coefficient
conf_interval = ols_model.conf_int(alpha=0.01).loc['arrivals']
print(f"99% Confidence Interval for arrivals' coefficient: {conf_interval.values}")

# Scatter plot and predictive line
plt.figure(figsize=(10, 6))
plt.scatter(training_data['arrivals'], training_data['vehicle_count'], alpha=0.5, label='Observed Data')
plt.plot(training_data['arrivals'], ols_model.predict(training_data), color='red', label='Predictive Line')
plt.title('Scatter Plot of Ridership vs Flight Arrivals')
plt.xlabel('Flight Arrivals')
plt.ylabel('Vehicle Count (Ridership)')
plt.legend()
plt.grid()
plt.show()
```

Q4: Incorporate Temporal Patterns

adding categorical variables for day of week and hour of the day.

From the visualization in task1Q3, you may see that both - taxi ridership as well as arrivals follow a somewhat periodic temporal pattern. Intuitively, this is true for most traffic flows following daily rhythms including rush hours, and also varying over the course of the week. Usually, we add dummy/categorical variables (Boolean variable: 1 for True and 0 for False) to encapsulate people's traveling patterns during different time periods.

Note that it would not make sense to add hour and dow as regular regressors as we can't anticipate their linear numeric impact. Instead, the expression C(.) or pd.get_dummies could be used in the regression formulae in order to treat those variables as categorical adding corresponding dummy variables to account for their possible discrete values.

Perform the regression of vehicle_count ridership against arrivals, lags, and temporal categorical variables.


```python
import statsmodels.formula.api as smf

# Build the regression formula including arrivals, lags, and temporal categorical variables
formula = (
    'vehicle_count ~ arrivals + ' +
    ' + '.join([f'lag{lag}' for lag in range(1, 13)]) +
    ' + C(DOW) + C(Hour)'
)

# Fit the OLS model
ols_model_with_temporal = smf.ols(formula=formula, data=training_data).fit()

# Print the model summary
print(ols_model_with_temporal.summary())

# Save coefficients and p-values to a CSV file
coefficients_df = pd.DataFrame({
    'Coefficient': ols_model_with_temporal.params,
    'P-Value': ols_model_with_temporal.pvalues
})
coefficients_df.to_csv("Model_Summary_With_Temporal.csv", index=True)
```


Q5. Perform feature selection for lag variables

As you may see not all the lag variables have a statistically significant impact on the regression. Maybe some of them are not really relevant? Try different amounts of lag variables m=0,1,...,12 (include m most recent arrival volumes/lags) using a loop for training the above regression over the training sample, report and plot the out-of-sample R2 over the validation sample as the function of m and pick the m which maximizes it. Evaluate the final regression over the test sample.


```python
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Store R^2 values for each m
validation_r2 = []

# Iterate over m = 0 to 12
for m in range(13):  # 0 to 12 lags
    # Construct formula with m lags
    lag_terms = ' + '.join([f'lag{lag}' for lag in range(1, m + 1)])
    formula = f'vehicle_count ~ arrivals + {lag_terms}' if m > 0 else 'vehicle_count ~ arrivals'

    # Train the model on the training data
    ols_model = smf.ols(formula=formula, data=training_data).fit()

    # Predict on the validation set and calculate R^2
    validation_predictions = ols_model.predict(validation_data)
    r2 = 1 - sum((validation_data['vehicle_count'] - validation_predictions)**2) / sum(
        (validation_data['vehicle_count'] - validation_data['vehicle_count'].mean())**2
    )
    validation_r2.append(r2)

# Plot R^2 vs m
plt.figure(figsize=(10, 6))
plt.plot(range(13), validation_r2, marker='o', linestyle='-', color='b')
plt.title('Out-of-Sample R^2 vs Number of Lag Variables (m)')
plt.xlabel('Number of Lag Variables (m)')
plt.ylabel('Validation R^2')
plt.grid()
plt.show()

# Find the optimal m
optimal_m = validation_r2.index(max(validation_r2))
print(f"Optimal number of lags (m): {optimal_m}")

# Train final regression model using training + validation data
lag_terms = ' + '.join([f'lag{lag}' for lag in range(1, optimal_m + 1)])
formula = f'vehicle_count ~ arrivals + {lag_terms}' if optimal_m > 0 else 'vehicle_count ~ arrivals'

final_model = smf.ols(formula=formula, data=pd.concat([training_data, validation_data])).fit()

# Evaluate the final model on the test set
test_predictions = final_model.predict(test_data)
test_r2 = 1 - sum((test_data['vehicle_count'] - test_predictions)**2) / sum(
    (test_data['vehicle_count'] - test_data['vehicle_count'].mean())**2
)

print(f"Test R^2 for the final model: {test_r2}")
```

Q2 (10 pts). Try K-means with different numbers of clusters k=2..7, report average Silhouette score for each. Which k is the "optimal" from Silhouette's standpoint?

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the normalized pivot table
data = normalized_pivot_table.values  # Use only the numerical data

# Store results
silhouette_scores = []

# Test k = 2 to 7
for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(data)

    # Compute silhouette score
    score = silhouette_score(data, cluster_labels)
    silhouette_scores.append(score)
    print(f"k={k}, Silhouette Score: {score:.4f}")

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 8), silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Score vs Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()

# Find the optimal k
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2  # Adding 2 because range starts from 2
print(f"Optimal number of clusters (k): {optimal_k}")
```

Optimal Number of Clusters (kk):

The optimal kk is the number of clusters that yields the highest Silhouette score. In this case:

    k=5k=5 has the highest Silhouette score (0.2377).
