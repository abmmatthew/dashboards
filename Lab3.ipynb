```python
def plot_loghist(x, bins): #introduce a function for plotting a log-scale histogram
  #it ensures log-scale binning and label on the original scale
    logbins = np.logspace(np.log10(x.min()),np.log10(x.max()),bins)
    plt.hist(x, bins=logbins)
    plt.xscale('log')
```

```python
#based on the above impose the data sanity index
sanityindex = (data.fare_amount >= 2.5) & (data.fare_amount <= 300) & \
    (data.trip_distance >= 0.3) & (data.trip_distance <= 100) & \
    (data.speed <= 80) & (data.speed >= 2) & \
    (data.trip_duration <= 100) & (data.trip_duration >= 1)
```

Spatial distribution of the taxi activity

Right now taxi zones are all of different size, while the activity heatmap is usually perceived as a density map. However the quantity above is affected by both - activity density as well as the size of the unit which is somewhat arbitrary. Normalizing activity by the area (trips per sq. mile) or by population (trips per capita) would be more appropriate


```python
tripsViz['ActivityDensity'] = tripsViz['N_trips'] / tripsViz['shape_area']
```


Explore temporal distribution of taxi ridership activity

Assess how the number of trips varies with the time of the day

make a column for pickup hour - 0-23 hrs (0 corresponds to 12am-1am window, 1 to 1am-2am window and so on...)


```python
data['pickUpHour'] = data['tpep_pickup_datetime'].dt.hour
data.head()
```

```python
#group trips by hour
tripsByHour = data[['pickUpHour', 'DOLocationID']].groupby(by='pickUpHour', as_index=False).count()
tripsByHour.rename(columns={'DOLocationID':'N_trips'}, inplace=True)
tripsByHour
```

```python
plt.figure(figsize=(15, 6))

plt.bar(tripsByHour.pickUpHour, tripsByHour.N_trips)
plt.xticks(tripsByHour.pickUpHour)
plt.ylabel('number of trips', fontsize=14)
plt.xlabel('hour of day', fontsize=14)
```


Explore average travel speed by hour of the day

In order to homogenize by traffic mileage we use the distance total/duration total instead of averaging individual trip speed


```python
speedByHour = data[['pickUpHour', 'trip_distance', 'trip_duration']].groupby(by='pickUpHour', as_index=False).sum()
speedByHour['speed'] = speedByHour['trip_distance'] / speedByHour['trip_duration'] * 60
speedByHour
```

```python
plt.figure(figsize=(15, 6))

plt.bar(speedByHour.pickUpHour, speedByHour.speed)
plt.xticks(speedByHour.pickUpHour)
plt.ylabel('average speed (mph)', fontsize=14)
plt.xlabel('hour of day', fontsize=14)
```

```python
# prompt: Visualize the spatial distribution (geopandas heatmap) of a) average travel speed (cumulative distance divided by cumulative time), and b) average trip distance by pick up location

# Calculate average travel speed and average trip distance by pick up location
avg_speed_by_zone = data.groupby('PULocationID').apply(lambda x: (x['trip_distance'].sum() / x['trip_duration'].sum()) * 60).reset_index(name='AvgSpeed')
avg_distance_by_zone = data.groupby('PULocationID')['trip_distance'].mean().reset_index(name='AvgDistance')

# Merge with the taxi zone GeoDataFrame
speed_viz = taxi.merge(avg_speed_by_zone, left_on='location_i', right_on='PULocationID', how='left')
distance_viz = taxi.merge(avg_distance_by_zone, left_on='location_i', right_on='PULocationID', how='left')

# Plot the heatmaps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))

speed_viz.plot(column='AvgSpeed', ax=ax1, legend=True, cmap='viridis', scheme='Quantiles', k=10,
              legend_kwds={'loc': 'upper left'}, missing_kwds={'color': 'lightgrey'})
ax1.set_title('Average Travel Speed by Pickup Location')

distance_viz.plot(column='AvgDistance', ax=ax2, legend=True, cmap='plasma', scheme='Quantiles', k=10,
              legend_kwds={'loc': 'upper left'}, missing_kwds={'color': 'lightgrey'})
ax2.set_title('Average Trip Distance by Pickup Location')

plt.show()
```

```python
# prompt: Visualize the timelines (bar plots) of the a) total number of trips, b) average speed (cumulative distance divided by cumulative time) by the day of the week (Sunday - Saturday)
# # make a column for day of week
# # group by for getting metrics, remember 'count()' for number of trips, 'sum()' for getting totals
# # make a bar plot

# Make a column for day of the week
data['dayOfWeek'] = data['tpep_pickup_datetime'].dt.dayofweek

# Group by day of the week for total number of trips
tripsByDayOfWeek = data[['dayOfWeek', 'DOLocationID']].groupby(by='dayOfWeek', as_index=False).count()
tripsByDayOfWeek.rename(columns={'DOLocationID':'N_trips'}, inplace=True)

# Group by day of the week for average speed
speedByDayOfWeek = data[['dayOfWeek', 'trip_distance', 'trip_duration']].groupby(by='dayOfWeek', as_index=False).sum()
speedByDayOfWeek['speed'] = speedByDayOfWeek['trip_distance'] / speedByDayOfWeek['trip_duration'] * 60

# Create bar plots
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.bar(tripsByDayOfWeek.dayOfWeek, tripsByDayOfWeek.N_trips)
plt.xticks(tripsByDayOfWeek.dayOfWeek, ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'])
plt.ylabel('number of trips', fontsize=14)
plt.xlabel('day of week', fontsize=14)
plt.title('Total Number of Trips by Day of Week')

plt.subplot(1, 2, 2)
plt.bar(speedByDayOfWeek.dayOfWeek, speedByDayOfWeek.speed)
plt.xticks(speedByDayOfWeek.dayOfWeek, ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'])
plt.ylabel('average speed (mph)', fontsize=14)
plt.xlabel('day of week', fontsize=14)
plt.title('Average Speed by Day of Week')

plt.tight_layout()
plt.show()
```
